In this notebook, I try out different thetas for the dissimilarity distance (the distance function from the paper "A Family of Dissimilarity Measures between Nodes Generalizing both the Shortest-Path and the Commute-time Distances" by Luh Yen, Marco Saerens et al.) and see what happens 

```python
n = 50
num_thetas = 10
large_number = 100
G = nx.erdos_renyi_graph(n, 0.15)

thetas = np.logspace(-1.3, 1, num_thetas)
```

```python
thetas
```

```python
pos = nx.spring_layout(G)
fig = plt.figure(figsize=(15, 10))
nx.draw_networkx_edges(G, pos=pos, alpha=0.5)
nx.draw_networkx_nodes(G, pos=pos, node_size=50, node_color='red')
plt.show()
```

```python
def get_distance_matrix_2(G, theta, large_number):
    n = G.number_of_nodes()
    M = np.zeros((n, n))
    for i in range(n):
        for j in range(i):
            M[i, j] = distances.diss_distance(i, j, theta, G, large_number)
    M_prime = M+np.transpose(M)
    return M_prime-np.diag(M_prime)
```

```python
for i in range(num_thetas):
    theta = thetas[i]
    print(f'theta = {theta}')
    mat = get_distance_matrix_2(G, theta, large_number)
    ax = sns.heatmap(mat, linewidth=0.5, vmax=22)
    plt.show()
    
```

```python
for i in range(num_thetas):
    theta = thetas[i]
    print(f'theta = {theta}')
    mat = get_distance_matrix_2(G, theta, large_number)
    ax = sns.heatmap(mat, linewidth=0.5)
    plt.show()
```

```python
mat1 = get_distance_matrix_2(G, theta=50, large_number)
mat2 = get_distance_matrix_2(G, theta=10, large_number)
mat1-mat2
```

The higher the theta, the lower the distances in general. Also, it seems that the distances seem to converge (see cell above).

However, the distances only seem to scale, not to change wildly (the patterns in the heatmaps look very similar for all the thetas)

From the paper:

First, it has a clear, intuitive, interpretation in terms of a biased random walk on the graph. Second, it nicely generalizes the shortest-path and the commute-time distances (also called the resistance dis- tance) by computing an intermediate dissimilarity depending on one parameter  (the inverse temperature). **When  is large, the dissimilarity reduces to the standard shortest-path distance while for  = 0, it reduces to the commute-time distance, also called the resistance distance [18, 30].** Third, the dissimilarity between a par- ticular node and all the other nodes can be computed efficiently by solving two linear systems of n equations where n is the number of nodes. On the other hand, all-pairs of nodes dissimilarity can be computed by inverting an n  n square matrix.
